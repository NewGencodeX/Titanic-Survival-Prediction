```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/akapo/Downloads/titanic/train.csv")
df
```

```python
df.describe()
```

```python
df.info()
```

```python
df['Survived'].value_counts().plot(kind='bar')
plt.title("Overall Survival Distribution")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Number of Passengers")
plt.xticks(rotation=0)

plt.savefig("../outputs/figures/survival_distribution.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
gender_survival = pd.crosstab(df['Sex'], df['Survived'])
gender_survival.plot(kind='bar')

plt.title("Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Passengers")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.xticks(rotation=0)

plt.savefig("../outputs/figures/survival_by_gender.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
class_survival = pd.crosstab(df['Pclass'], df['Survived'])
class_survival.plot(kind='bar')

plt.title("Survival by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Number of Passengers")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.xticks(rotation=0)

plt.savefig("../outputs/figures/survival_by_class.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
survived_age = df[df['Survived'] == 1]['Age'].dropna()
not_survived_age = df[df['Survived'] == 0]['Age'].dropna()

plt.hist(not_survived_age, bins=30, alpha=0.7, label='Did Not Survive')
plt.hist(survived_age, bins=30, alpha=0.7, label='Survived')

plt.title("Age Distribution by Survival")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.legend()

plt.savefig("../outputs/figures/survival_by_age.png", dpi=300, bbox_inches="tight")
plt.show()
```
